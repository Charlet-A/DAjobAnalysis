```python
df.columns = df.columns.str.strip()

df.columns = df.columns.str.replace(" ","_")

df.columns
```

```python
df.describe(include='all')
```

```python
df['Industry'] = df['Industry'].fillna('Unknown')
```

```python
df['Sector'] = df['Sector'].fillna('Unknown')
```

```python
df['Revenue'] = df['Revenue'].fillna('Not Specified')
```

```python
df['Rating'].unique()
```

```python
df['Company_Name'].head()
```

```python
df['Company_Name'] = df['Company_Name'].str.split('\n').str[0]
```

```python
df['Salary_Estimate'].head()
```

```python
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('$','',regex=False)

df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('K','',regex=False)
```

```python
df['Min_Salary'] = df['Salary_Estimate'].str.extract(r'(\d+)')
```

```python
df['Max_Salary'] = df['Salary_Estimate'].str.extract(r'-(\d+)')
```

```python
df['Min_Salary'] = pd.to_numeric(df['Min_Salary'],errors='coerce')

df['Max_Salary'] = pd.to_numeric(df['Max_Salary'],errors='coerce')
```

```python
df['Avg_Salary'] = (df['Min_Salary'] + df['Max_Salary']) / 2
```

```python
df[['Min_Salary','Max_Salary','Avg_Salary']].head()
```

```python
df['Company_Age'] = 2026 - df['Founded']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
top_companies = df['Company_Name'].value_counts().head(10)

plt.figure(figsize=(12,6))
top_companies.sort_values().plot(kind='barh')

plt.title('Top 10 Hiring Companies')
plt.xlabel('Number of Job Postings')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

```python
top_locations = df['Location'].value_counts().head(10)

plt.figure(figsize=(12,6))
top_locations.plot(kind='bar')

plt.title('Top 10 Hiring Locations')
plt.xlabel('Location')
plt.ylabel('Job Count')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['Avg_Salary'],
    bins=25,
    kde=True
)

plt.title('Salary Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    y=df['Rating']
)

plt.title('Company Rating Distribution')
plt.show()
```

```python
industry = df['Industry'].value_counts().head(8)

plt.figure(figsize=(8,8))

plt.pie(
    industry.values,
    labels=industry.index,
    autopct='%1.1f%%'
)

centre = plt.Circle((0,0),0.70)

fig = plt.gcf()

fig.gca().add_artist(centre)

plt.title('Industry Share')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Rating',
    y='Avg_Salary'
)

plt.title('Rating vs Salary')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x='Company_Age',
    y='Avg_Salary'
)

plt.title('Company Age vs Salary')

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    y='Size',
    data=df,
    order=df['Size'].value_counts().index
)

plt.title('Company Size Distribution')

plt.show()
```

```python
top_ind = df['Industry'].value_counts().head(6).index

plt.figure(figsize=(14,6))

sns.violinplot(
    data=df[df['Industry'].isin(top_ind)],
    x='Industry',
    y='Avg_Salary'
)

plt.xticks(rotation=45)

plt.title('Salary Distribution by Industry')

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True
)

plt.title('Correlation Heatmap')

plt.show()
```

```python
location_salary = (
    df.groupby('Location')['Avg_Salary']
      .mean()
      .sort_values(ascending=False)
      .head(15)
)

plt.figure(figsize=(12,6))

plt.hlines(
    y=location_salary.index,
    xmin=0,
    xmax=location_salary.values
)

plt.plot(
    location_salary.values,
    location_salary.index,
    "o"
)

plt.title('Top Salary Locations')

plt.show()
```
